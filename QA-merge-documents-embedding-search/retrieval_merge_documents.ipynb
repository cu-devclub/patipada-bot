## Retrieval

```python
from llama_index.core import StorageContext,load_index_from_storage
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
import nest_asyncio
nest_asyncio.apply()

embed_model = HuggingFaceEmbedding(model_name="intfloat/multilingual-e5-large",device='cpu')

load_index = load_index_from_storage(StorageContext.from_defaults(persist_dir='C:/Users/sukpi/QA-vector-database/localvecstore-merge-documents'),embed_model=embed_model)
```

```python
retrieval_engine = load_index.as_retriever(similarity_top_k = 10)
response = retrieval_engine.retrieve("ความโกรธ")
print('-------------RAG-------------')

for i in response :
    print(i)
```
